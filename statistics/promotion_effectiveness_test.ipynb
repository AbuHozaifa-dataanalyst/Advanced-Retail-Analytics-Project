# Promotion Effectiveness Test

## Business Question
Do promotional campaigns significantly increase sales compared to normal periods?

Retail companies use statistical testing to determine whether promotions actually improve revenue.

## Prepare Data

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

sales = pd.read_csv("../data/processed/Sales.csv")

promo_sales = sales[sales['Discount'] > 0]['Total_Amount']

normal_sales = sales[sales['Discount'] == 0]['Total_Amount']
```

## Hypothesis

Null Hypothesis (H0):
Promotions do not affect sales revenue.

Alternative Hypothesis (H1):
Promotions increase sales revenue.

## T-Test

```python
t_stat, p_value = stats.ttest_ind(promo_sales, normal_sales)

print("T-statistic:", t_stat)
print("P-value:", p_value)

# Compare average sales
means = [promo_sales.mean(), normal_sales.mean()]
labels = ['Promotion Sales', 'Normal Sales']

plt.bar(labels, means)
plt.title("Promotion vs Normal Sales Comparison")
plt.ylabel("Average Revenue")

# Save image for GitHub
plt.savefig("../images/promotion_test_result.png")

plt.show()
```

### Promotion Sales Comparison

The chart below compares average sales during promotional periods versus normal periods.
This visualization supports the statistical test results.

## Interpretation

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: Promotions significantly impact sales")
else:
    print("Fail to reject the null hypothesis")
```

## Business Recommendation

Business Recommendation

If promotions significantly increase revenue, the company should expand promotional campaigns during high-demand periods.

If not, the company should redesign promotional strategies.
